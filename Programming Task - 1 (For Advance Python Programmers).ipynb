## Q1.Running Sum of 1d Array

```python
class Solution:
    def runningSum(Self, arr):
        for i in range(1, len(arr)):
            arr[i] = arr[i] + arr[i - 1]
        return arr

data = eval(input())

solver = Solution()
result = solver.runningSum(data)
print(result)
```

## Q2.Shuffle the Array

```python
def shuffle_array(nums, n):
    result = []
    for i in range(n):
        result.append(nums[i])
        result.append(nums[i + n])

    return result

arr = eval(input())

n = int(input())

shuffled = shuffle_array(arr, n)
print(shuffled)
```

## Q3.Kids with the Greatest Number of Candies

```python
class Solution:
    def kidswithCandies(Self, Candies, extraCandies):
        max_candies = max(candies)
        result = []
        for candy in candies:
            result.append(candy + extraCandies >= max_candies)

        return result

candies = eval(input(""))
extraCandies = int(input(""))

solver = Solution()
print(solver.kidswithCandies(candies, extraCandies))
```
