# python If-Else

```python
n = int(input().strip())

if n % 2 == 1:
    print("Weird")
else:
    if n%2==0 and 2<=n<=5:
       print("Not Weird")
    elif n%2==0 and 6<=n<=20:
        print("Weird")
    else:  
        print("Not Weird")
```

# write a function

```python
year=int(input())
def is_leap(year):
    if year%400==0:
        return True
    if year%100==0:
        return False
    if year%4==0:
        return True
    return False
print(is_leap(year))
```
